```python
tiles = [f'{i+1}萬' for i in range(9)] + [f'{i+1}筒' for i in range(9)] + [f'{i+1}索' for i in range(9)] + list('東南西北白発中')
names = ['ドラ牌'] + [f'捨牌{i}' for i in range(21)] + [f'ツモ切り{i}' for i in range(21)] + ['リーチ巡目'] + [f'{t}待' for t in tiles]

df = pd.read_csv('data/ver2.csv', names=names, dtype={f: np.uint8 for f in names})
df = df[::-1]
# df.to_feather('data/ver2.feather')
```

```python
l = []
prev = np.zeros(44, dtype=np.uint8)
for i, row in df.iterrows():
    row = row.to_numpy()
    if not (prev[0] == row[0] and prev[43] == row[43] and np.all(prev[44:]==row[44:]) and np.count_nonzero(prev[1:43]!=row[1:43]) <= 2):
        l.append(i)
    prev = row
```

```python
reduced = df.loc[l]
```
